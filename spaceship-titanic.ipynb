```python
train_data = pd.read_csv("/kaggle/input/spaceship-titanic/train.csv")
train_data.head()
test_data = pd.read_csv('/kaggle/input/spaceship-titanic/test.csv')
test_data.head()
```

```python
y = train_data.Transported

features = ['CryoSleep','Age','RoomService','VRDeck']
x = train_data[features]
y.head()
x.describe()
```

```python
from sklearn.ensemble import RandomForestRegressor

#X = x.dropna()
#Y = y.loc[X.index]

from sklearn.impute import SimpleImputer

imp = SimpleImputer(strategy = 'median')
X_clean = imp.fit_transform(x)
#Y_clean = imp.transform(train_data[features])

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

#train_X,val_X,train_Y,val_Y = train_test_split(X_clean,Y_clean,random_state = 1)

model = RandomForestRegressor(random_state = 1,max_leaf_nodes = 2500)
model.fit(X_clean,y)
#predictions = model.predict()
#mae = mean_absolute_error(predictions,val_Y)
#print(2300,mae)

x_test = test_data[features]
X_test_clean = imp.transform(x_test)
predictions = model.predict(X_test_clean)
predictions_list = []

for i in range(len(predictions)):
    if predictions[i] < 0.5:
        predictions_list.append(False)
    else:
        predictions_list.append(True)

output = pd.DataFrame({'PassengerId':test_data.PassengerId,'Transported':predictions_list})
output.to_csv('submission.csv',index = False)
print(output)


```
