```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# Set pandas option to avoid downcasting warnings
pd.set_option('future.no_silent_downcasting', True)

# Load data
train_data = pd.read_csv("/kaggle/input/spaceship-titanic/train.csv")
test_data = pd.read_csv('/kaggle/input/spaceship-titanic/test.csv')

# Prepare features and target
features = ['HomePlanet', 'CryoSleep', 'Destination', 'Age', 'VIP', 
            'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

x = train_data[features]
y = train_data.Transported
x_test = test_data[features]

# Clean preprocessing function
def clean_preprocess(df):
    df_processed = df.copy()
    
    # Categorical columns
    categorical_cols = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP']
    
    for col in categorical_cols:
        # Fill missing values using recommended approach
        if df_processed[col].isna().any():
            mode_val = df_processed[col].mode().iloc[0] if not df_processed[col].mode().empty else 'Unknown'
            filled_series = df_processed[col].fillna(mode_val)
            # Use infer_objects as recommended
            df_processed[col] = filled_series.infer_objects(copy=False)
        
        # Encode categorical variables
        le = LabelEncoder()
        df_processed[col] = le.fit_transform(df_processed[col].astype(str))
    
    # Numerical columns
    numerical_cols = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
    for col in numerical_cols:
        if df_processed[col].isna().any():
            median_val = df_processed[col].median()
            df_processed[col] = df_processed[col].fillna(median_val)
    
    return df_processed

# Alternative: Use a dictionary approach for fillna (most modern way)
def modern_preprocess(df):
    df_processed = df.copy()
    
    # Create fill dictionaries
    fill_dict = {}
    
    # Categorical columns - fill with mode
    categorical_cols = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP']
    for col in categorical_cols:
        mode_val = df_processed[col].mode().iloc[0] if not df_processed[col].mode().empty else 'Unknown'
        fill_dict[col] = mode_val
    
    # Numerical columns - fill with median
    numerical_cols = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
    for col in numerical_cols:
        fill_dict[col] = df_processed[col].median()
    
    # Fill all missing values at once
    df_processed = df_processed.fillna(fill_dict)
    
    # Encode categorical variables
    for col in categorical_cols:
        le = LabelEncoder()
        df_processed[col] = le.fit_transform(df_processed[col].astype(str))
    
    return df_processed

# Use the modern preprocessing approach
X_processed = modern_preprocess(x)
X_test_processed = modern_preprocess(x_test)

print("Processed training data shape:", X_processed.shape)
print("Processed test data shape:", X_test_processed.shape)

# Train model
model = RandomForestRegressor(random_state=1, n_estimators=100)
model.fit(X_processed, y)

# Make predictions
predictions = model.predict(X_test_processed)
predictions_list = [True if pred >= 0.5 else False for pred in predictions]

# Create submission
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Transported': predictions_list})
output.to_csv('submission.csv', index=False)
print("\nSubmission created successfully!")
print(output.head())
```
